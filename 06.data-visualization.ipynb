1. Line plot

```python
# create line plot
import pandas as pd
from matplotlib import pyplot as plt

```

```python
series = read_csv('daily-minimum-temperatures.csv', parse_dates=True, squeeze=True, index_col=0)
print(series.describe())
series.plot()
plt.show()
```

```python
# use dot instead of line
series.plot(style='.k')
plt.show()
```

```python
# Grouper
# Grouped by year
years = series.groupby(pd.Grouper(freq='A'))
years.plot()
```
