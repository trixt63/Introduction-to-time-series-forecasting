# Decompose Time Series Data
* Additive Model:

```y(t) = Level + Trend + Seasonality + Noise```

* Multiplicative Model

```y(t) = Level * Trend * Seasonality * Noise```

````python
from random import randrange
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
````

````python
# Create a time series comprised of a linearly increasing trend from 1 to 99 and add some random noise
series = [i+randrange(10) for i in range(1, 100)]
series = pd.Series(series)
print(type(series))
````

````python


# Decompose addictive
result = seasonal_decompose(series, model="additive", period = 1)
print(type(result))
result.plot()
plt.show()
````

We can see that there are no seasonality, while trend is incrementalm and residual is random.

````python
# Decompose mulplicative
series = [i**2 for i in range(1, 100)]
result =seasonal_decompose(series, model="mulplicative", period=1)
result.plot()
plt.show()
````
