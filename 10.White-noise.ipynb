# A Gentle Introduction to White Noise
A time series is white noise if:
- Mean = 0
- Std is constant with time
- No auto-correlation (correlation between lag)
Why does it matter:
- Predictability
- Model Diagnostics

```python
from random import gauss, seed
import pandas as pd
from matplotlib import pyplot as plt

# seed random number generator
random.seed(1)
# create white noise series
series = [gauss(0.0, 1.0) for i in range(1000)]
series = pd.Series(series)
```

```python
print(series.describe())
```

```python
series[:200].plot()
plt.show()
```

```python
series[:200].hist()
plt.show()
# We will see a bell-shape
```

```python
# autocorrelation
pd.plotting.autocorrelation_plot(series)
# Show no obvious autocorrelation
```
